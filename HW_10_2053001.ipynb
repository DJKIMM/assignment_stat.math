##4. 자연수 13과 14가 각각 소수인지를 python을 이용하여 판별해 보시오. 

#### method 1) if-else문 사용

```python
x = 13
for i in range(2, 13):
  if x % i == 0:
    print("%d는 소수가 아닙니다." % x)
    break
  else:
    print("%d는 소수입니다." % x)
    break
```

```python
x = 14
for i in range(2, 14):
  if x % i == 0:
    print("%d는 소수가 아닙니다." % x)
    break
  else:
    print("%d는 소수입니다." % x)
    break
```

method 2) 함수 사용

```python
# x가 1보다 큰 정수라고 가정
def prime(x):
  if x == 2:
    return "%d는 소수입니다." % x
  else:
    for i in range(2, x): # 2부터 (x - 1)로 나누어지지 않는 숫자는 소수
      if x % i == 0:
        return "%d는 소수가 아닙니다." % x
      else:
        return "%d는 소수입니다." % x
```

```python
# 13일 때 확인
prime(13)
```

```python
# 14일 때 확인
prime(14)
```
