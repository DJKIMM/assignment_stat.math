# Python problem 1: 
## Write  python  function  to  calculate  the  inner  product  using  matrix  multiplication where  inputs  are  Numpy  column  vectors $u$ and $v$.  Then,  try  you  function  with Problem 6 in Chapter 4

```python
import numpy as np
```

##Answer

```python
def inner(u, v): # 매개변수 u, v는 column vector
  return u.T @ v
```

### 결과값 확인

```python
# 6 - (1)
u = np.array([-1, 3]).T
v = np.array([2, 4]).T
print(inner(u, v))
```

```python
# 6 - (2)
u = np.array([-1, 1, 4]).T
v = np.array([3, 2, -2]).T
print(inner(u, v))
```

```python
# 6 - (3)
u = np.array([1, -1, 1, 4]).T
v = np.array([1, 1, 2, -2]).T
print(inner(u, v))
```

```python
# 6 - (4)
u = np.array([-1, 1, 2, 4]).T
v = np.array([1, 0, 2, -2]).T
print(inner(u, v))
```

# Python problem 3: 
## Write python function to return the unit length vector from the input Numpy vectors $u$. (Here, $u$ is assumed to be a non-zero vecto.) Specifically, your return should be $1/||u|| * u$

```python
def unit(u): # 매개변수는 vector u
  return 1/(np.sqrt(np.inner(u, u))) * u
```

```python
# 확인
v = np.array([-1, 3])
print(unit(v))
```

# Python problem 4.
## In $R^3$, you want to calculate orthogonal projection of $a = (a_1, a_2, a_3)$ in $R^3$ onto $W = <(u_1, u_2, u_3), (v_1, v_2, v_3)>$. Use the Python code to calculate the orthogonal projection of $a = (1, 0, 2)$ in $R^3$ onto $W = <(1, 1, 0), (1, 0, 1)>$.

```python
# Calculate projw(a) = p
  # projw(a) = A @ inv((A.T) @ A) @ A.T @ a
A = np.array([[1, 1], [1, 0], [0, 1]])
a = np.array([[1], [0], [2]])
p = A @ np.linalg.inv((A.T) @ A) @ A.T @ a
print(p)
```

```python
x1 = np.array([[1, 1, 0, 0]]).T
x2 = np.array([[0, 1, 1, 0]]).T
x3 = np.array([[0, 0, 1, 1]]).T
x4 = np.array([[0, 0, 0, 1]]).T
```

```python
v1 = x1
v1
```

```python
# v2 = x2 - proj<v1>x2
v2 = x2 - ((x2.T @ v1) / (v1.T @ v1)) * v1
v2
```

```python
# v3 = x3 - proj<v1, v2>x3
v3 = x3 - ((x3.T @ v1) / (v1.T @ v1)) * v1 -  ((x3.T @ v2) / (v2.T @ v2)) * v2
v3
```

```python
# v4 = x4 - proj<v1, v2, v3>x4
v4 = x4 - ((x4.T @ v1) / (v1.T @ v1)) * v1 - ((x4.T @ v2) / (v2.T @ v2)) * v2 - ((x4.T @ v3) / (v3.T @ v3)) * v3
v4
```
