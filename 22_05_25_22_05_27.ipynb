* while문 강제로 빠져나가기 : break

```python
coffee = 10 # 자판기가 팔 수 있는 커피는 maximum이 10개
money = 300 # money가 바뀌지 않으므로 무한루프
while money:
  print("커피")
  coffee = coffee - 1
  print("남은 커피의 양은 %d 개입니다." % coffee)
  if coffee == 0:
    print("Sold Out")
    break # while 문에서 나옴
```

```python
coffee = 10
while True: # 무한루프
  # input은 character로 바뀌어서 나오기 때문에 int로 형변환
  money = int(input("돈을 넣어주세요: "))
  if money == 300:
    print("커피")
    coffee = coffee - 1
  elif money > 300:
    print("커피, 거스름돈 %d" % (money - 300))
    coffee = coffee - 1
  else:
    print("돈 돌려주기")
  print("남은 커피 %d" % coffee)
  if coffee == 0:
    print("Sold Out!")
    break

# 300 넣으면 커피, 남은 커피 출력 - 커피 개수 차감
# 400 넣으면 커피, 거스름돈 출력 - 커피 개수 차감
# 200 넣으면 돈 돌려주기 출력, 커피 개수 차감하지 않음
# 커피를 모두 소진하면 커피, 남은 커피 0, Sold Out! 출력 
```

* continue : 건너뜀

```python
a = 0
while a < 10:
  a = a + 1
  # a를 2로 나누었을 때 나머지가 0이면 건너뛰고 while문으로 돌아감
  if a % 2 == 0: continue
  print(a)
```

* "*" 이용 : 받아들이는 매개변수가 몇 개인지 명확하지 않을 때 이용

```python
def add_many(*args):
  result = 0
  for i in args: # args: 넣어준 숫자
    result = result + i
  return(result)

add_many(1, 2, 3)
```

```python
def add_mul(choice, *args): # 개수가 정확한 것을 먼저 써줘야 함
  if choice == "add":
    result = 0
    for i in args:
      result = result + i
  elif choice == "mul":
    result = 1
    for i in args:
      result = result * i
  return(result)
```

```python
add_mul("add", 1, 2, 3, 4, 5)
```

```python
add_mul("mul", 1, 2, 3, 4, 5)
```

## 파일 open해서 읽기
* readline : 한 줄 씩 읽기
* readlines : 모두 읽기
* read : 하나의 문자열로 출력

```python
# readline 사용
f = open("c:\\TEMP-2022\\newfile-1.txt", 'r')
line = f.readline()
print(line)

# line = f.readline()
# print(line)
# 입력하면 첫 줄부터 순서대로 읽음
# 모든 줄을 읽으면 출력하지 않음

f.close()
```

```python
# while문 사용
f = open("c:\\TEMP-2022\\newfile-1.txt", 'r')
while True: # 무한 루프
  line = f.readline()
  if not line: break # 비어 있는 line이 나오면 while문장이 끝남 -> 출력하지 않음
  print(line)
f.close()
```

```python
# readlines 사용
f = open("c:\\TEMP-2022\\newfile-1.txt", 'r')
lines = f.readlines()
for line in lines:
  print(line) # lines 출력해보면 \n이 있어서 한 줄씩 띄워진 상태로 출력
f.close()
```

```python
# read 사용
f = open("c:\\TEMP-2022\\newfile-1.txt", 'r')
data = f.read() # data 출력해보면 하나의 문자열로 출력
print(data) # data 확인해보면 \n이 있어서 한 줄씩 띄워진 상태로 출력
f.close()
```
